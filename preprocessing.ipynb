```
def func(row):
    if row['DepTime'] %100!=0:
        k=row['DepTime']//100 + 1 if row['DepTime'] %100>30 else 0 
        return k
    else:
        return row['DepTime']//100

r['DepTime'] = r.apply(func, axis=1)
```

```
l=['ATL','CLT','DEN','DFW','EWR','IAH','JFK','LAS','LAX','MCO','MIA','ORD','PHX','SEA','SFO']
t = pd.DataFrame()
for apt in l:
  temp = r[r.Origin == apt]
  t = pd.concat([t,temp],axis = 0)
```

```
import os
f=0
needed = ['2016-1.json',
          '2016-2.json',
          '2016-3.json',
          '2016-4.json',
          '2016-5.json',
          '2016-6.json',
          '2016-7.json',
          '2016-8.json',
          '2016-9.json',
          '2016-10.json',
          '2016-11.json',
          '2016-12.json',
          '2017-1.json',
          '2017-2.json',
          '2017-3.json',
          '2017-4.json',
          '2017-5.json',
          '2017-6.json',
          '2017-7.json',
          '2017-8.json',
          '2017-9.json',
          '2017-10.json',
          '2017-11.json',
          '2017-12.json']
w=pd.DataFrame()
m=1
for i in os.listdir('/content/gdrive/My Drive/ML/weather') :
  currdir = '/content/gdrive/My Drive/ML/weather/'+i
  temp=pd.DataFrame()
  for j in needed:
    filename = '/content/gdrive/My Drive/ML/weather/'+i+'/'+j
    w1= pd.read_json(filename)
    day_cnt=1
    for p in range(len(w1['data']['weather'])):
      day = pd.DataFrame()
      for h in range(24):
        d=pd.DataFrame(w1['data']['weather'][p]['hourly'][h])
        d['DepTime']=h
        d['Month']=(j.split('-'))[1].split('.')[0]
        d['year']=(j.split('-'))[0]
        d['DayofMonth']=day_cnt
        day=pd.concat([day,d],axis=0,ignore_index=True)
      temp=pd.concat([temp,day],axis=0,ignore_index=True)
      day_cnt+=1
  temp['Origin']=i    
  w=pd.concat([w,temp],axis=0,ignore_index=True)
  print(i+' '+j)
  print(w.shape)
```

```
w.to_csv('/content/gdrive/My Drive/ML/weatherDataset.csv', sep='\t', encoding='utf-8')
```
