```
w=pd.read_csv('/content/gdrive/My Drive/ML/weatherDataset.csv', sep='\t', encoding='utf-8')
```

```
w =  w.rename( columns={"DepTime": "CRSDepTime"})
```

```
def func(row):
    return row['CRSDepTime']//1
w['CRSDepTime'] = w.apply(func, axis=1)
```

```
f=pd.read_csv('/content/gdrive/My Drive/ML/flightDataset.csv', sep='\t', encoding='utf-8')
```

```
def func(row):
    k = row['DepTime']//100
    if row['DepTime']%100 >30:
      k+=1
    return k
f['DepTime'] = f.apply(func, axis=1)
```

```
w =  w.rename( columns={"CRSDepTime":"DepTime"})
final_data = pd.merge(f,w,how='outer',on=['DepTime','Origin','Month','DayofMonth','year'])
```
